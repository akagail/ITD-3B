```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings 
import folium
warnings.filterwarnings('ignore')
```

```python


# Load the dataset with the correct encoding
df = pd.read_csv('datasets\\Global_terrorist.csv', encoding='ISO-8859-1')

```

```python
df
```

<h1>Filtering Data</h1>

```python
filter1 = df[df['region_txt'] == 'Middle East & North Africa']
```

```python
filter2 = filter1[filter1['iyear'] > 2000]
```

```python
filter3 = filter2[filter2['attacktype1_txt'] == 'Bombing/Explosion']
```

```python
filter3
```

<h1>Data Aggregation</h1>

```python
filter3.columns
```

```python
coordinates = filter3.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
```

```python
coordinates
```

<h1>Use of Folium for GeoSpatial</h1>

```python
import folium
from folium.plugins import HeatMap
```

<h1>Converting Pandas dataframe to python list</h1>

```python
heatmap_data = coordinates[['latitude', 'longitude', 'incident_count']].values.tolist()
```

<h1>Creating a Base Map</h1>

```python
my_map = folium.Map(location = [30, 40], zoom_start = 4)
```

<h1>Add heatmap layer</h1>

```python
HeatMap(heatmap_data).add_to(my_map)
```

```python
my_map.save('terrorism_bombing.html')
```

```python
my_map
```

```python
filter1_i1 = df[df['country_txt'] == 'Philippines']
```

```python
bar = filter1_i1['attacktype1_txt'].value_counts()
```

```python
filter2_i1 = filter1_i1[filter1_i1['attacktype1_txt'] == 'Armed Assault']
```

```python
ax = bar.plot(kind='bar')

for x in ax.containers:
    ax.bar_label(x, padding=3)
    
plt.gcf().set_size_inches(19, 8)
plt.title('Value Counts of the Attacks in the Philippines')
plt.show()
```
